```python
#import all necessary module
import langchain
import langchain_huggingface
import langchain_community
from sentence_transformers import SentenceTransformer
import faiss
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm=HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation",
    temperature=0.5,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_ACCESS_TOKEN"]
)
```

```python
model=ChatHuggingFace(llm=llm)
```

```python
#create chat template and dynamic prompt
chat_template=ChatPromptTemplate([
    ('system','You are a helpful {domain} expert'),
    ('human','Tell me about what is {topic}')
])
prompt=chat_template.invoke({'domain':'football','topic':'penalty'})
```

```python
#put this peompt in the llm
response=model.invoke(prompt)
print(response.content)
```
